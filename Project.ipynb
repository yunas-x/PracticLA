```python
model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

```python
import numpy as np

def normalized(a, axis=-1, order=2):
    l2 = np.atleast_1d(np.linalg.norm(a, order, axis))
    l2[l2==0] = 1
    return a / np.expand_dims(l2, axis)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from numpy.linalg import norm
```

```python
sentence_embeddings_PI = {k: model.encode(v) for k,v in PI_competences.items()}
```

```python
sentence_embeddings_BI = {k: model.encode(v) for k,v in BI_competences.items()}
```

```python
PI = pd.read_excel("gdrive/MyDrive/Colab Notebooks/ПИ_DATASET.xlsx")
BI = pd.read_excel("gdrive/MyDrive/Colab Notebooks/БИ_DATASET.xlsx")
```

```python
PI = PI.set_index('Предмет').T.to_dict()
BI = BI.set_index('Предмет').T.to_dict()
```

```python
def compare_competences(course_1, course_2) -> int:
  comps1 = course_1["Компетенции"].split(", ")
  comps2 = course_2["Компетенции"].split(", ")
  vecs1 = [sentence_embeddings_PI.get(c, 0) for c in comps1]
  vecs2 = [sentence_embeddings_BI.get(c, 0) for c in comps2]
  mean_vec1 = normalized(sum(vecs1))
  mean_vec2 = normalized(sum(vecs2))
  return norm(mean_vec1 - mean_vec2, ord=2),  cosine_similarity(mean_vec1, mean_vec2)

def compare_names(name_1, name_2) -> int:
  emb1 = normalized(model.encode(name_1))
  emb2 = normalized(model.encode(name_2))
  return norm(emb1 - emb2, ord=2), cosine_similarity(emb1, emb2)
```

```python
compare_competences(PI["Алгебра"], BI["Алгебра и геометрия"])

names1 = []
names2 = []
compare1 = []
compare2 = []

for name1, info1 in PI.items():
  for name2, info2 in BI.items():
    names1.append(name1)
    names2.append(name2)
    compare1.append(compare_competences(info1, info2)[1][0][0])
    compare2.append(compare_names(name1, name2)[1][0][0])
```

```python
df = pd.DataFrame(data={
    "name1": names1,
    "name2": names2,
    "comp_comp": compare1,
    "comp_name": compare2
})
```
